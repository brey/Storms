## JTWC

```python
url = 'https://metoc.ndbc.noaa.gov/RSSFeeds-portlet/img/jtwc/jtwc.rss'
```

```python
print readxml(url,'lxml')
```

```python
jtwc=feedparser.parse(url)
```

```python
data=jtwc['entries']
for k in range(len(data)):
    if 'tcw' in data[k].summary : txt = (data[k].summary)
```

```python
txt.split('href')
```

```python
elp = [re.sub('<[^<]+>', "", elem) for elem in txt.split('href=')]
```

```python
bul = [elem for elem in elp if 'tcw' in elem][0]
```

```python
bul = bul.split()[0].strip()
```

```python
bul=bul.encode('utf8').replace('"', '')
```

```python
det = urllib.urlopen(bul).read()
```

```python
print det
```

```python
data =[line.strip() for line in det.splitlines() if ('T' is line.strip()[0]) & ('QD' in line.strip()[-2:])]
```

```python
data = [re.sub(' ',',',elem) for elem in data]
```

```python
data = [v.split(',') for v in data]
```

```python
[l[i:i + 13] for i in xrange(4, len(l), 13) for l in data]
    
```

```python
df = pd.DataFrame(data)[0].str[1:-1].str.split(',', expand=True)
```
